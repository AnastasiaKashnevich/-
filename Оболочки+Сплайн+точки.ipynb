```python
def calculationCoefficient(t):
    return [((1-t)**2)/2,(-2*t**2+2*t+1)/2,(t**2)/2]
    
def generationCoefficient(stepCount):
    Coefficient = list()
    for i in range(0,stepCount):
        Coefficient.append(calculationCoefficient(float(i/stepCount)))
    return Coefficient

def intermediateValue(stepCount,points,ValuesCoefficient):
    Values = np.empty([3,stepCount])
    for i in range(0,stepCount):
        for j in range(0,3):
            Values[j][i] = 0
            for l in range (0,3):
                Values[j][i] += ValuesCoefficient[i][l]*points.selectElement(l).getCoordinate(j+1)
    return Values

def plotGraph(VectorPoints,ValueCoefficient,Closed,MainPoint):
    fig = plt.figure()
    ax = Axes3D(fig)
    if MainPoint == True:
        ax.plot(VectorPoints.getCoordinates(1), VectorPoints.getCoordinates(2), VectorPoints.getCoordinates(3), "b")
    if Closed == False:
        Values = np.empty([3,10*(a.Quantity()-2)+2])
        for i in range(0,a.Quantity()-2):
            tmp = Points()
            for j in range(0,3):
                tmp.addElement(VectorPoints.selectElement(i+j))
            PointGraph = intermediateValue(10,tmp,ValueCoefficient)
            for j in range(0,3):
                for k in range(0,10):
                    Values[j][10*i+k+1]=PointGraph[j][k]
        for j in range(0,3):
            Values[j][0] = VectorPoints.selectElement(0).getCoordinate(j+1)
            Values[j][10*(a.Quantity()-2)+1] = VectorPoints.selectElement(a.Quantity()-1).getCoordinate(j+1)
    else:
        Values = np.empty([3,10*(a.Quantity())+1])
        for i in range(0,a.Quantity()):
            tmp = Points()
            if i == a.Quantity()-2:
                tmp.addElement(VectorPoints.selectElement(i))
                tmp.addElement(VectorPoints.selectElement(i+1))
                tmp.addElement(VectorPoints.selectElement(0))
            elif i == a.Quantity()-1:
                tmp.addElement(VectorPoints.selectElement(i))
                tmp.addElement(VectorPoints.selectElement(0))
                tmp.addElement(VectorPoints.selectElement(1))
            else:
                for j in range(0,3):
                    tmp.addElement(VectorPoints.selectElement(i+j))
            PointGraph = intermediateValue(10,tmp,ValueCoefficient)
            for j in range(0,3):
                for k in range(0,10):
                    Values[j][10*i+k]=PointGraph[j][k]
        for i in range(0,3):
            Values[i][10*(a.Quantity())] = Values[i][0]
    ax.plot(Values[0], Values[1], Values[2], "k")
    plt.show()
    
```

```python
#Точки
%matplotlib
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
#from scipy.spatial import ConvexHull
import matplotlib
import random

class Point:
    def __init__(self,x,y,z):
        self.x=x
        self.y=y
        self.z=z
        
    def getCoordinate(self,index):
        if index == 1: return self.x
        elif index == 2: return self.y
        else: return self.z
    
class Points:
    def __init__(self):
        self.Vector = list()
        
    def clearElements(self):
        self.Vector.clear()
        
    def Quantity(self):
        return len(self.Vector)
        
    def addElement(self,point):
        self.Vector.append(point)
        
    def selectElement(self,index):
        return self.Vector[index]
    
    def deleteElement(self,index):
        self.Vector.pop(index)
        
    def getCoordinates(self,index):
        Values = np.empty([1,len(self.Vector)])
        for i in range (0,len(self.Vector)):
            Values[0][i] = self.Vector[i].getCoordinate(index)
        return Values[0]
```
